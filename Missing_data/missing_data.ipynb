```python
import pandas as pd

#Create df
weather_steps = pd.read_csv("weather_steps.csv")
weather_steps = weather_steps.drop(weather_steps.columns[0], axis=1)
display(weather_steps)
```

```python
print(weather_steps.isna().sum())
#remove temp_min_6h and weather_code, because they contain more than 80% missing values    
weather_steps = weather_steps.drop(["temp_min_6h", "weather_code"] , axis=1)           
```

```python
# Convert the 'time' column to datetime
weather_steps['creationDate'] = pd.to_datetime(weather_steps['creationDate'], format='%Y-%m-%d %H:%M')

# Interpolate the missing values
weather_steps['windspeed_avg_hour'] = weather_steps['windspeed_avg_hour'].interpolate()
weather_steps['windspeed_avg_10min'] = weather_steps['windspeed_avg_10min'].interpolate()
weather_steps['max_wind_gust'] = weather_steps['max_wind_gust'].interpolate()
weather_steps['temp_celsius'] = weather_steps['temp_celsius'].interpolate()
weather_steps['temp_dewpoint'] = weather_steps['temp_dewpoint'].interpolate()
weather_steps['sunshine_duration'] = weather_steps['sunshine_duration'].interpolate()
weather_steps['glob_radiation'] = weather_steps['glob_radiation'].interpolate()
weather_steps['precipitation_duration'] = weather_steps['precipitation_duration'].interpolate()
weather_steps['precipitation_amount_hourly'] = weather_steps['precipitation_amount_hourly'].interpolate()
weather_steps['air_pressure'] = weather_steps['air_pressure'].interpolate()
weather_steps['horizontal_visibility'] = weather_steps['horizontal_visibility'].interpolate()
weather_steps['cloud_cover'] = weather_steps['cloud_cover'].interpolate()
weather_steps['relative_humidity'] = weather_steps['relative_humidity'].interpolate()
weather_steps['fog'] = weather_steps['fog'].interpolate()
weather_steps['rain'] = weather_steps['rain'].interpolate()
weather_steps['snow'] = weather_steps['snow'].interpolate()
weather_steps['thunder'] = weather_steps['thunder'].interpolate()
weather_steps['ice_formation'] = weather_steps['ice_formation'].interpolate()
```
